## Problem Statment:

### The goal of the program is to classify the tweets into a good tweet or a racist/sexist tweet. The training data already consists of the classified tweets into good and bad with the labels 0 and 1 respectively.

### It is easier to predict rather than making use of human resource to spend their time identify which tweet belongs to what category. Thus, the problem is to predict the labels of the tweets in the training data set. 


### Import the necesary libraries

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from autoviml.Auto_NLP import Auto_NLP

import warnings
warnings.simplefilter(action='ignore')
```

### Loading both the train and test data of tweets

```python
train = pd.read_csv('train_tweet.csv')
test = pd.read_csv('test_tweet.csv')
```

```python
train.head()
```

```python
test.head()
```

### Counting the nummber of tweets that are good, and bad which are labelled as 0 and 1's respectively 

```python
train.label.value_counts()
```

### Splitting the data into train test split

```python
train1, test = train_test_split(train, test_size=0.2)
```

### Assigning the input variable data to input feature variable and the label to a target variable

```python
input_feature, target = "tweet", "label"
```

### Fitting the data on the AutoNLP model and setting teh required features to fine tune the model

```python
train_x, test_x, final, predicted= Auto_NLP(input_feature, train1, test,target,
                                            score_type="balanced_accuracy",
                                            top_num_features=100,
                                            modeltype="Classification",
                                            verbose=2,
                                            build_model=True)
```

### Predict the labels for the x_test data, that was split in the train data itself, with the Auto NLP model, that is built

```python
final.predict(test_x[input_feature])
```

### Read the test_tweet data

```python
testing = pd.read_csv('test_tweet.csv')
```

### Predict the labels for the final test data

```python
final.predict(testing[input_feature])
```

### Read the sample output file

```python
sample = pd.read_csv('sample_submission_gfvA5FD.csv')
```

### Creating a new column as label, on the sample dataframe and assigning it the predicted tweet lables of the test data

```python
sample['label'] = final.predict(testing[input_feature])
```
